```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# Paths
dir_non = '../images_dataSAT/class_0_non_agri/'
dir_agri = '../images_dataSAT/class_1_agri/'

# ✅ Task 1: determine the shape of one image
sample_path = os.path.join(dir_non, os.listdir(dir_non)[0])
img = Image.open(sample_path)
print("Image size:", img.size, "Mode:", img.mode)

# ✅ Task 2: display first 4 non-agri images
files_non = sorted(os.listdir(dir_non))[:4]
for f in files_non:
    img = Image.open(os.path.join(dir_non, f))
    plt.imshow(img)
    plt.title(f)
    plt.axis('off')
    plt.show()

# ✅ Task 3: create list of agri image paths (sorted)
agri_images_paths = sorted([os.path.join(dir_agri, f) for f in os.listdir(dir_agri)])
print("List of agri image paths:")
for p in agri_images_paths:
    print(p)

# ✅ Task 4: number of agri images
print("Number of agricultural images:", len(agri_images_paths))

# Optional: display them
for f in agri_images_paths[:4]:
    img = Image.open(f)
    plt.imshow(img)
    plt.title(f)
    plt.axis('off')
    plt.show()
```
